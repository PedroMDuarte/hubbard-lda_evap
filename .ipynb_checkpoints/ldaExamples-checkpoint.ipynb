## Default 

```python

import scubic
import lda

pot = scubic.sc(allGR=3.621)
test = lda.lda(potential = pot, a_s=650., Temperature=0.12, extents=50., \
               globalMu='halfMott', SN=1.60, verbose=True, ignoreLowT=True)
fig = lda.plotLine( test , line_direction='111', extents=50., ignoreLowT=True) 
fig.savefig('Examples_figures/default.png',dpi=300)
```

## Harmonic

```python

import scubic
import lda

wIR=500.
wGR=120.
s0 = 7.
g0 = -4.2

pot = scubic.sc(allIR=s0,allGR=g0, allIRw=wIR, allGRw=wGR)
test = lda.lda(potential = pot, a_s=650., Temperature=0.12, extents=50., \
               globalMu='halfMott', SN=1.60, verbose=True, ignoreLowT=True)
print test.pot.TrapFreqs()

fig = lda.plotLine( test , line_direction='111', extents=50., ignoreLowT=True) 
fig.savefig('Examples_figures/harmonic.png',dpi=300)

print test.getRadius()
```

## Mathy's prescription for the compensating potential is 

## $g_{0} = \beta (s_{0})^{\alpha^{2}} $

```python
import scubic
import lda

# Mathy Fig. 7 
beta = 0.3 
alpha = 1.16
s0 = 8. 
print beta * ( s0 **(alpha**2.))

a_s = 0.1 * 20113/2.
print a_s


pot = scubic.sc(allIR=8.0, allGR=4.924, allIRw=47., allGRw=40.5)
test = lda.lda(potential = pot, a_s=1005., globalMu=0.95, verbose=True, ignore_errors = True)
fig = lda.plotMathy( test , line_direction='111', ax1ylim=(-2.0,0.)) 
fig.savefig('Examples_figures/defaultMathy_Fig7.png',dpi=300)
```
