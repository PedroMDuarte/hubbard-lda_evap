```python
x = np.linspace(-25, 25, 240)

def f(x, U):
    return np.piecewise(x, [x <= 0., \
                            np.logical_and( x>0 , x <= U/2 ), \
                            np.logical_and( x>U/2, x<=U ),\
                            x> U  ], \
                        [lambda x: np.zeros_like(x), \
                         lambda x: x/(U/2.), \
                         lambda x: 2.-x/(U/2.), lambda x:np.zeros_like(x) ])

def g(x, U):
    return np.piecewise( x, [ x <= U/2., x>U/2. ],
             [ lambda x: np.exp( -np.power( ( x-U/2.)/(U/2.), 2.)), 
               lambda x: np.ones_like(x) ] ) 

plt.plot( x, f(x, 10.) )
plt.plot( x, g(x, 10.) ) 
plt.ylim(0., 1.1)
```
