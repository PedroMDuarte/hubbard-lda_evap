## Understanding how can $\eta=1$ for a large beam waist with moderate atom number 

```python
import scubic
import lda

pot = scubic.sc(allGR=3.642)
test = lda.lda(potential = pot, a_s=650., Natoms=3.5e5, muBrent=(-1.,6.), verbose=True)
fig = lda.plotLine( test , line_direction='111') 
fig.savefig('Examples_figures/default.png',dpi=300)
```
