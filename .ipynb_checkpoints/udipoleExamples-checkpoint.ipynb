### Show some examples of usage for the potential class

# Apparatus3 Cross beam dipole trap

```python
from udipole import *

odt = potential([ GaussBeam(mW=38000., waists=(70.,70.), axis=(np.pi/2,30.*np.pi/180.)),
                  GaussBeam(mW=38000., waists=(70.,70.), axis=(np.pi/2,15.*np.pi/180.)) ])

fig = plt.figure( figsize = (8., 8.) ) 
gs = matplotlib.gridspec.GridSpec( 3,2, wspace=0.2) 

# Make a list with three perpendicular directions which 
# will define the three surface cuts 
perp = [(np.pi/2., 0.), (np.pi/2., -np.pi/2.), (0., -1.*np.pi/2.) ]

# Iterate to plot the three surface cuts
yMin = 1e16
yMax = -1e16 
Ims = []
for i in range(3):
    ax0 = fig.add_subplot( gs[i,0], projection='3d')
    ax1 = fig.add_subplot( gs[i,1])  
    
    T0, T1, X, Y, Z = surfcut_points( normal = perp[i], extents=800., ax0=ax0 ) 
    
    EVAL = odt.evalpotential(X,Y,Z)
    im = ax1.pcolormesh( T0, T1, EVAL, \
                   cmap=plt.get_cmap('jet') ) 
    plt.axes( ax1 )  
    cbar = plt.colorbar(im)
    cbar.set_label( odt.unitlabel, rotation=0)  
    
    ymin = EVAL.min()
    ymax = EVAL.max()
    
    Ims.append(im) 
    if ymin < yMin : yMin = ymin
    if ymax > yMax : yMax = ymax 
    
for im in Ims:
    im.set_clim( vmin=yMin, vmax=yMax) 
        
```

# Triangular lattice

```python
from udipole import *

lattTri = potential([ LatticeBeam( axis=(np.pi/2,0) ),
                     LatticeBeam( axis=(np.pi/2,2*np.pi*(1./3.)) ),
                     LatticeBeam( axis=(np.pi/2,2*np.pi*(2./3.)) )], units=('$E_{R}$', 1/Erecoil(1.064,6)) )
plot3surface( lattTri, extents = 80.)
```
