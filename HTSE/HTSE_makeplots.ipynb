Example to make a chemical potential cut at fixed U, starting from a filling of 1 particle per site.   Compare with Fig 5. in Fuchs et. al. PRL 106, 030401 (2011).   This plot is useful to see at what temperature does the HTSE break down:  a lot of wiggles start appearing in the entropy.  

```python
import htse 

def trap_fixed_U( U ):
    mu_ = np.linspace( -20, U/2, 100)
    U_  = np.ones_like( mu_ ) * U
    return np.transpose(np.vstack(( U_, mu_)))

for Temp in [1.0, 1.6, 1.8, 2.6, 4.6, 10.8 ]:
    U = 20. 
    mu = np.linspace( -20., U, 100) 
    z = htse.htse_entr(Temp, 1., mu, U, ignoreLowT=True)
    plt.plot( mu, z ) 
```

Make a plot of the density as a function of chemical potential for various interaction strenghts and temperatures

```python
from matplotlib import rc
rc('font',**{'family':'serif'})
rc('text', usetex=True)

def mu_cut( U, T ):
    mu_ = np.linspace( -14, 32, 100)
    U_  = np.ones_like( mu_ ) * U
    return mu_, htse.htse_dens( T, 1., mu_, U_ )

def plotCompare( ax, Uval=8. ):
    for Temp in [1.8,3.6,5.4]:
        x, y = mu_cut( Uval, Temp) 
        ax.plot( x, y,lw=2, label = "$T/t=%.1f$"%Temp) 
            

        
fig = plt.figure(figsize=(5.6,2.2))
gs = matplotlib.gridspec.GridSpec( 1,3, wspace=0.08, left=0.12, right=0.97, bottom=0.20, top=0.9)

ax0 = fig.add_subplot( gs[0,0])
ax1 = fig.add_subplot( gs[0,1])
ax2 = fig.add_subplot( gs[0,2])

label = '$\mathrm{atoms\ per\ site}$'
ax0.set_ylabel(label,fontsize=14,rotation=90,labelpad=12)

axs = [ax0,ax1,ax2]
for ax in axs:
    ax.set_xlabel("$\mu/t$",fontsize=12)
    ax.set_xlim( -12., 32.)
    ax.set_ylim( -0.05, 2.05 )
    ax.grid()
    ax.xaxis.set_minor_locator(matplotlib.ticker.MultipleLocator(2))
    ax.grid(which='minor')
    
for i,Uval in enumerate([8., 24., 36.]):
    plotCompare(axs[i], Uval=Uval)
    axs[i].set_title('$U/t = %d$'%Uval, fontsize=10, ha='center')
    
for ax in [ax1,ax2]:
    for tick in ax.get_yticklabels():
        tick.set_visible(False)

bbox=(1.0,0.00)
loc='lower right'
ax2.legend( bbox_to_anchor=bbox, \
            loc=loc, numpoints=1, \
             prop={'size':8}, handlelength=1.1, handletextpad=0.5 )

#gs.tight_layout(fig, rect=[0.,0.,0.9,0.92])
fig.savefig('figures/HTSE_density_U.png',dpi=180)
```
